# Datos de incidencia delictiva

Datos obtenidos de: https://datos.gob.mx/busca/dataset/incidencia-delictiva-del-fuero-comun-a-nivel-municipal

```python
import pandas as pd

df = pd.read_csv('data-raw/IDM_NM_4.csv',sep=';')
print(df.columns)
df.tail()
```

```python
df = df.groupby(['Clave_Ent', 'Entidad', 'Cve. Municipio', 'Municipio', 'Bien jurídico afectado', 
                 'Tipo de delito', 'Subtipo de delito', 'Modalidad']) \
                .agg({'TOTAL ESTADO':'sum'}).reset_index()
df['Clave_Ent'] = df['Clave_Ent'].apply(lambda x: str(x).zfill(2))
df['Cve. Municipio'] = df['Cve. Municipio'].apply(lambda x: str(x).zfill(5))
df.to_csv('results/IncidenciaDelictivaNacional-MunicipioDelito.csv', index=False)
df.head(15)
```

```python
pivot = df.pivot_table(df, index=['Clave_Ent', 'Entidad', 'Cve. Municipio', 'Municipio'],
                       columns=['Tipo de delito', 'Subtipo de delito','Modalidad'], 
                       aggfunc='sum')
pivot.reset_index(inplace=True)
pivot.columns = ['Clave_Ent','Entidad','Cve. Municipio','Municipio'] + [','.join(col[1:]).strip() for col in pivot.columns.values[4:] ]
pivot.to_csv('results/IncidenciaDelictivaNacional-MunicipioDelito-pivot.csv', index=False)
pivot.tail()
```
