# Panteones

* Registro de información "panteinf"

```python
import pandas as pd
from IPython.display import display
import matplotlib.pyplot as plt

f_dic = "diccionario_de_datos\diccionario_de_datos_{}_ricpzmp2020.csv"
f_con = "conjunto_datos\{}_ricpzmp2020.csv"
f_cat = "catalogos\{}_ricpzmp2020.csv"

keys = ['zonametr_a', 'entidad_a', 'ubicageo_a', 'nombre1', 'localida_a']
```

```python
def sustituir_catalogo(df, ndir, nvar):
    s = ndir+f_cat.format(nvar)
    dic = pd.read_csv(s, encoding='latin1')
    dic = dict(zip(dic[nvar], dic['descrip']))
    df[nvar].replace(dic, inplace=True)
    return df

def read_and_clean_csv(ndir, nset):
    # Cargar
    s = ndir+f_con.format(nset)
    print(s)
    df = pd.read_csv(s, encoding='latin1')
    # Datos del diccionario
    s = ndir+f_dic.format(nset)
    print(s)
    dic_col = pd.read_csv(s, encoding='latin1')
    dic_col = dict(zip(dic_col['COLUMNA'], dic_col['DESCRIPCION']))
    # Sustituir valores de catálogo
    for col in df.columns:
        if col not in keys: #catálogo
            try: df = sustituir_catalogo(df, ndir, col)
            except FileNotFoundError: print("\tNo hay catálogo para", col)
            # columna del diccionario
            df = df.rename(columns={col:dic_col[col]})
    return df
```

## Datos de contacto de panteones

```python
ndir = "panteones_ricpzmp2020_csv/"
nset = "panteon"
nvar = "rensnn1"

print( "PANTEONES" )
df_pan = read_and_clean_csv(ndir, nset)
print(df_pan.shape)
df_pan.tail()
```

## Datos de fosas

```python
ndir = "panteones_operacion_ricpzmp2020_csv/"
nset = "pantefos"

print( "FOSAS" )
df_fos = read_and_clean_csv(ndir, nset)
print(df_fos.shape)
df_fos.tail()
```

Agregar datos de contacto


NOTA:  Hay 3958 pero solo 1431 en la tabla de Fosas Comunes. 
Solo se agrega datos alos panteones de "pantefos", que suponemos son los que contestaron la pregunta de fosas comunes.


```python
df_fos = pd.merge(df_fos, df_pan, how="left", on=keys)
df_fos
```

Panteones donde no hay información de fosas en el conteo del INEGI

```python
df_desconocido = df_fos[df_fos['Existencia de fosas comunes'].isin(["Información no disponible", "No se sabe"])]

for col in keys:
    print(col)
    try: df_desconocido = sustituir_catalogo(df_desconocido, ndir, col)
    except: print("\tNo hay catálogo para", col)

df_desconocido.to_csv("PanteonesDatosFosasDesconocido.csv")
df_desconocido.tail()
```

Panteones con fosas comunes (respondieron Sí)

```python
df_fos = df_fos[df_fos['Existencia de fosas comunes'].isin(["Sí"])]
df_fos
```

Agregar los datos de información

```python
ndir = "panteones_operacion_ricpzmp2020_csv/"
nset = "panteinf"

print( "INFORMACION" )
df_inf = read_and_clean_csv(ndir, nset)
print(df_inf.shape)
df_inf.tail()
```
